```python
def plot(num):
    # country
    df_location = ['LSE','LSE','LSE','LSE','Brussels','XETRA','Paris',
                    'Swiss','Milan','LSE','LSE','LSE','LSE','NYSE',
                    'NYSE','NYSE','NYSE','NYSE','NYSE','NYSE','NYSE','NYSE','NYSE',
                    'NYSE','Paris','XETRA','Swiss','XETRA','Japan',
                    'Japan','Japan','ASX','NYSE','Toronto','LSE','LSE',
                    'LSE','LSE','LSE','LSE','LSE','LSE','Amsterdam',
                    'Johannesburg','Johannesburg','Oslo','Oslo','Milan','Amsterdam',
                    'Copenhagen','Swiss','Helsinki','Copenhagen','IOB','Swiss',
                    'Toronto'
    ]
    ticker = sentiment_analysis_df_with_price['ticker'].unique()
    df_country = pd.DataFrame({'Stock_market':df_location,'ticker':ticker})
    df_merge = sentiment_analysis_df_with_price.merge(df_country, how='inner', on='ticker')
    #df_merge

    # df_2 = sentiment_analysis_df_with_price.drop(['company_name','ticker','file_name','Close','D0'], axis = 1)
    # df_2['year']=df_2['date'].apply(lambda r: r[:4])

    df = df_merge.drop(['company_name','file_name','Close','D0'], axis = 1)
    df['year']=df['date'].apply(lambda r: r[:4])
    df['id']=df['Group']+'-'+df['Stock_market']
    #df = df.drop(df[df.ticker == 'QBE.AX'].index)
    ctg_list = df['id'].unique().tolist()
    # day_list = sentiment_analysis_df_with_price.columns[8:]
    df

    df_2 = df.copy()
    df_motor = df_2[df_2['id']==ctg_list[num]].sort_values(by='year')
    print(ctg_list[num])
    print(df_motor['ticker'].unique())

    fig = make_subplots(rows=2, cols=1)

    year = df_motor['year'].unique()
    median_list_sentiment = []

    for y in year :
        y_value = df_motor[df_motor['year']==y]['WA_sentiment']
        median_val = np.median(y_value)
        median_list_sentiment.append(median_val)
        fig.add_trace(go.Box(y=y_value, name=y, marker_color='royalblue'),
                        row=1, col=1)
    fig.add_trace(
        go.Scatter(
            x=year,
            y=median_list_sentiment
        ),
        row=1, col=1
        )

    # second plot
    median_list_prc = []
    for y in year :
        y_value = df_motor[df_motor['year']==y]['D-1']
        median_val = np.median(y_value)
        median_list_prc.append(median_val)
        fig.add_trace(go.Box(y=y_value, name=y, marker_color='royalblue'),
                        row=2, col=1)
    fig.add_trace(
        go.Scatter(
            x=year,
            y=median_list_prc
        ),
        row=2, col=1
        )

    fig.update_layout(height=500, width=1000)

    fig.show()
```
